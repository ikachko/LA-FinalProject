```python
import librosa
import librosa.display as libdisp
import IPython.display as ipd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
!ls guitars/
```

```python
vocal_path = 'vocals/vocal lead.wav'
guitar_path = 'guitars/Gtr 1.wav'

x_voc, sr_voc = librosa.load(vocal_path)
x_g1, sr_g1 = librosa.load(guitar_path)
```

```python
x_all = x_voc * 0.4 + x_g1 * 0.6

librosa.output.write_wav('guitar_vocal.wav', x_all, sr_voc)
```

```python
sr_voc, sr_g1
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(14, 10))

axs[0].set_title('Lead vocal')
libdisp.waveplot(x_voc, sr=sr_voc, ax=axs[0])

axs[1].set_title('Guitar 1')
libdisp.waveplot(x_g1, sr=sr_g1, ax=axs[1])
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(14, 10))

X_v = librosa.stft(x_voc)
Xdb_v = librosa.amplitude_to_db(abs(X_v))

X_g1 = librosa.stft(x_g1)
Xdb_g1 = librosa.amplitude_to_db(abs(X_g1))

axs[0].set_title('Lead vocal spectrogram')
libdisp.specshow(Xdb_v, sr=sr_voc, x_axis='time', y_axis='hz', ax=axs[0])

axs[1].set_title('Guitar 1 spectrogram')
libdisp.specshow(Xdb_g1, sr=sr_g1, x_axis='time', y_axis='hz', ax=axs[1])
```
